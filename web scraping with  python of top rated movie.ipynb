```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/chart/top/")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
movie_names=[]
imdb_ratings=[]
year_of_release=[]
```

```python
titles=soup.find('tbody',class_='lister-list').find_all('td',class_='titleColumn')
titles[0:99]
```

```python
for i in titles:
    
    full_title=""
    for j in i.find_all('a'):
        full_title=full_title+""+j.get_text()
        print(full_title)
        titles.append(full_title)
```

```python
imdb_ratings=soup.find_all('td',class_="ratingColumn imdbRating")
imdb_ratings[0:99]
```

```python
year_of_release=soup.find_all('span',class_="secondaryInfo")
year_of_release[0:99]
```

```python
movie_names=soup.find_all('td',class_="watchlistColumn")
movie_names[0:99]
```

```python
print(len(movie_names),len(imdb_ratings),len(year_of_release))
```

```python
import pandas as pd
df=pd.DataFrame({})
df['movie']=movie_names[0:99]
df['imdb_ratings']=imdb_ratings[0:99]
df['year_of_release']=year_of_release[0:99]
```
