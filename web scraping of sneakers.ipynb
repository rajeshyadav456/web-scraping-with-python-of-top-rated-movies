```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.flipkart.com/mens-footwear/casual-shoes/sneakers~type/pr?sid=osp%2Ccil%2Ce1f')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
Brand=soup.find_all('div',class_="_2WkVRV")
Brand[0:100]
```

```python
price=soup.find_all('div',class_="_3I9_wc")
price[0:100]
```

```python
discount=soup.find_all('div',class_="_30jeq3")
discount[0:100]
```

```python
production_Description=soup.find_all('a',class_="IRpwTa _2-ICcC")
production_Description[0:100]
```

```python
print(len(Brand),len(production_Description),len(price),len(discount))
```
