```python
from  bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
book_name=[]
author_name=[]
book_review=[]
```

```python
book_name=soup.find_all('h4',class_="italic")
book_name[0:5]
```

```python
author_name=soup.find_all('p',class_="sans bold")
author_name[0:5]
```

```python
book_review=soup.find_all('li')
book_review[0:5]
```

```python
print(len(book_name),len(author_name),len(book_review))
```

```python
import pandas as pd 
df=pd.DataFrame({})
df['book_name']=book_name[0:5]
df['author_name']=author_name[0:5]
df['book_review']=book_review[0:5]
```
