```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get('https://www.ebmnews.com/?s=news++of+2019')
```

```python
r.content
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
headline=soup.find_all('h2',class_="title")
headline[0:100]
```

```python
print(len(headline))
```

```python
author=soup.find_all('i',class_="post-author author")
author[0:100]
```

```python
print(len(author))
```

```python
date=soup.find_all('time',class_="post-published updated")
date[0:100]
```

```python
print(len(date))
```

```python
description=soup.find_all('div',class_="post-summary")
description[0:100]
```

```python
print(len(description))
```

```python
verticle=soup.find_all('a')
verticle[0:100]
```

```python
print(len(verticle))
```

```python
import pandas as pd
df=pd.DataFrame({})
df['verticle']=verticle[:10]
df['headline']=headline[:10]
df['author']=author[:10]
df['date']=date[:10]
df['descrition']=description[:10]
```
