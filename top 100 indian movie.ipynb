```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
movie_names=[]
imdb_rating=[]
year_of_release=[]
```

```python
title=soup.find_all('tbody',class_="lister-list")
title[0:100]
```

```python
movie_names=soup.find_all('td',class_="watchlistColumn")
movie_names[0:100]
```

```python
imdb_rating=soup.find_all('td',class_="ratingColumn imdbRating")
imdb_rating[0:100]
```

```python
year_of_release=soup.find_all('span',class_="secondaryInfo")
year_of_release[0:100]
```

```python
print(len(movie_names),len(imdb_rating),len(year_of_release))
```

```python
import pandas as pd
df=pd.DataFrame({})
df['movie_names']=movie_names[0:100]
df['imdb_rating']=imdb_rating[0:100]
df['year_of_release']=year_of_release[0:100]
```
