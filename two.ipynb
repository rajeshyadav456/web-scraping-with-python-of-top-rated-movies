```python
from bs4 import BeautifulSoup
import requests
```

```python
r1=requests.get('https://www.ebmnews.com/category/sports/')
```

```python
r1.text
```

```python
soup=BeautifulSoup(r1.text,'html.parser')
```

```python
soup.title
```

```python
soup.title.string
```

```python
print(soup.prettify())
```

```python
headline=[]
date=[]
authors_name=[]
description=[]
```

```python
headline=soup.find_all('a',class_="post-url post-title")
print(headline)
```

```python
for i in headline:
    print(i.text.split)
    
```

```python
date=soup.find_all('time',class_="post-published updated")
print(date)
```

```python
for i in date:
    print(i.text.split('[]'))
```

```python
authors_name=soup.find_all('a',class_="post-author-a")
print(authors_name)
```

```python
for i in authors_name:
    print(i.text.split)
```

```python
description=soup.find_all('div',class_="post-summary")
print(description)
```

```python
for i in description:
    print(i.text.split)
```

```python
import pandas as pd
df=pd.DataFrame({})
df['headline']=headline
df['date']=date
df['authors_name']=authors_name
df['description']=description
```
